```python
import pandas as pd            # For data manipulation
import numpy as np             # For numerical operations
import matplotlib.pyplot as plt # For basic plotting
import seaborn as sns          # For advanced visualization
import requests                # For web scraping
from bs4 import BeautifulSoup  # For parsing HTML
```

```python
# 1. Connect to the site
url = "http://quotes.toscrape.com"
response = requests.get(url)

# 2. Parse the content
soup = BeautifulSoup(response.text, 'html.parser')

# 3. Extract data (Find all quotes)
quotes = [q.text for q in soup.find_all('span', class_='text')]
authors = [a.text for a in soup.find_all('small', class_='author')]

# 4. Create the DataFrame
df = pd.DataFrame({'Quote': quotes, 'Author': authors})
```

```python
# 1. Set the figure size (wider for horizontal labels)
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")

# 2. Switch axes: Put 'Author' on x and count on y
# We use 'xticks' rotation so the names don't overlap
sns.countplot(x='Author', data=df, palette='viridis', order=df['Author'].value_counts().index)

# 3. Rotate the author names so they are readable
plt.xticks(rotation=45)

# 4. Add titles and labels
plt.title('Distribution of Quotes per Author (Vertical View)', fontsize=15)
plt.xlabel('Author Name', fontsize=12)
plt.ylabel('Number of Quotes', fontsize=12)

# 5. Show the plot
plt.show()
```
